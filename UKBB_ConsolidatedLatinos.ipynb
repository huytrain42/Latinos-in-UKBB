```python
#Read in UKBB Demographics
ukbb = pd.read_csv(r"C:\Users\Huy\Desktop\Dataset\ukb_demographics.txt", sep='\t')
ukbb
```

```python
#Read in Latinos Country of Origin
latinos_country = pd.read_csv(r"C:\Users\Huy\Desktop\Dataset\latinos_countryOfOrigin.txt", sep='\t')
latinos_country
```

```python
#Get Latino cohort by matching EID with UKBB Demographic and Latino Country of Origin
ukbb_new = pd.merge(latinos_country, ukbb, how='inner', on=['eid'])
ukbb_new['EthnicGroup'] = ukbb_new['EthnicGroup'].replace(['White', 'OtherEthnicity', 'Mixed', 'Chinese', 'Asian', 'Black'],'Latino')
ukbb_new
```

```python
#Retrieving the ancestry data
ancestry = pd.read_csv(r"C:\Users\Huy\Desktop\Dataset\latino_admixture_k4.txt", sep='\t')
ancestry = ancestry.dropna()
ancestry = ancestry.astype({"IID": 'int64'})
ancestry
```

```python
#Latino cohort
temp1 = pd.merge(ukbb_new, ancestry, left_on='eid', right_on='IID', how='left')
temp1
```

```python
#Read in disparate disease list
disparate = pd.read_csv(r"C:\Users\Huy\Desktop\Dataset\disparate_diseases.txt", sep='\t')
disparate
```

```python
folder = os.listdir("C:\\Users\\Huy\\Desktop\\Dataset\\phecode_cohorts")
list_top = []
for disease in disparate.head(10).Disease: #Go through only top 10 disparate
    list_top.append(disease)
ukbb_update = pd.DataFrame(columns = [])
for file in folder:
    if file in list_top:
        disease_df = pd.read_csv(r"C:\\Users\\Huy\\Desktop\\Dataset\\phecode_cohorts\\" + file, sep='\t')
        disease_df["Disease"] = file #Get name of disease
        temp2 = pd.merge(temp1, disease_df, how='inner', on=['eid']) #Merge disease file and Latino cohort
        ukbb_update = ukbb_update.append(temp2) #Add to empty df and repeat
ukbb_update
```

```python
#Clean-up and removal of unnecessary columns
del ukbb_update["CountryCode"]
del ukbb_update["Country_x"]
del ukbb_update["Continent"]
del ukbb_update["AssessmentCenter"]
del ukbb_update["IID"]
del ukbb_update["FamID"]
del ukbb_update["SuperPop"]
del ukbb_update["Population"]
ukbb_update = ukbb_update.rename(columns={'eid':"EID"})
ukbb_update = ukbb_update.rename(columns={'Country_y':"Country"})
ukbb_update = ukbb_update[["EID", "Age", "Sex", "TownsendIndex", "EthnicBackground", "EthnicGroup", "Status", "Disease", "EastAsian", "African", "European", "Native", "Country"]]
ukbb_update
```

```python
temp_df2 = pd.DataFrame(columns = [])
disease_list_done = []
for disease in disparate.head(10).Disease: #Only look at top 10 disparate
    if disease in ["426.91_Cardiac_pacemaker_in_situ.txt"]:
        if disease not in disease_list_done:
            ageDict = {} #Keep track of age for every disease
            sexDict = {} #Keep track of sex for every disease
            Total = 0 #Keep track of total for every disease
            temp_df1 = ukbb_update[ukbb_update["Disease"] == disease] #Create tempory df of disease you want to look at
            for num in temp_df1.Age: #Go through participant age and add to dictionary
                if num not in ageDict:
                    ageDict[num] = 1
                else:
                    ageDict[num] = ageDict[num] + 1
            for gender in temp_df1.Sex: #Go through participant gender and add to dictionary
                if gender not in sexDict:
                    sexDict[gender] = 1
                    Total += 1
                else:
                    sexDict[gender] = sexDict[gender] + 1
                    Total += 1
            for index, row in temp_df1.iterrows(): #Go through every row in tempory df of the certain disease
                row["AgeFraction"] = (ageDict[row.Age] / Total) #Create column for age fraction
                row["SexFraction"] = (sexDict[row.Sex] / Total) #Create column for sex fraction
                temp_df2 = temp_df2.append(row) #Add new row to empty df and repeat
temp_df2
```

```python
#Reorder of columns
del temp_df2["Country"]
new_df = temp_df2[["EID", "EastAsian", "African", "European", "Native", "EthnicBackground", "EthnicGroup", "TownsendIndex", "Age", "AgeFraction", "Sex", "SexFraction", "Disease", "Status"]]
new_df["EID"] = new_df["EID"].astype(int)
new_df["Age"] = new_df["Age"].astype(int)
new_df["Status"] = new_df["Status"].astype(int)
new_df
```

```python
new_df.to_csv('top10_disparate_disease_consolidated_latino_treatment.txt', index=False, sep='\t')
```
